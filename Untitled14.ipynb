```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
data=pd.read_csv("CC GENERAL.csv", sep=",")
data.head
data.info()
```

```python
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(),inplace=True)
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(),inplace=True)
data = data.iloc[: , 1:]
data_scaled=data.apply(zscore)
data_scaled=normalize(data_scaled)
M=PCA(n_components=0.95)
M.fit_transform(data_scaled)
data.info()
```

```python
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
data['Agglo_clusters_labels']=clust_labels
fig =plt.figure()
ax = fig.add_subplot(111)
agglomerative=pd.DataFrame(clust_labels)
scatter = ax.scatter (data ['PURCHASES'] , data ["BALANCE"] , c= agglomerative[0], s=50)
plt.colorbar(scatter)
```

```python
S=[]
K=range(1,30)
for k in K:
  km=KMeans(n_clusters=k)
  km=km.fit(data)
  S.append(km.inertia_)
plt.plot(K,S)
plt.show
```
