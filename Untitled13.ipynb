```python
import pandas as pd 
data ["PAYMENTS"], data ["CREDIT_LIMIT"], data ["PURCHASES_TRX"], data ["CASH_ADVANCE_TRX"], data ["CASH_ADVANCE_FREQUENCY"], data ["PURCHASES_INSTALLMENTS_FREQUENCY"], data ["PURCHASES_INSTALLMENTS_FREQUENCY"], data ["ONEOFF_PURCHASES_FREQUENCY"], data ["PURCHASES_FREQUENCY"], data ["CASH_ADVANCE"], data ["INSTALLMENTS_PURCHASES"], data ["ONEOFF_PURCHASES"], data ["PURCHASES"], data ["BALANCE_FREQUENCY"], data ["BALANCE"], data ["MINIMUM_PAYMENTS"]
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn import metrics
from sklearn.tree import plot_tree, export_text
dataset=pd.read_csv("titanic-passengers.csv", sep=";")
#Preprocessing our data
def preprocess_data(new_data):
    new_data['Age'].fillna(new_data['Age'].mean(),inplace=True)
    new_data.replace({'Sex':{'male': 1,'female':0}},inplace=True)
    new_data['Cabin']=new_data.Cabin.fillna('G6')
    new_data.replace({'Survived':{'Yes': 1,'No':0}},inplace=True)
    new_data.replace({'Embarked':{'Q': 2,"S": 1,"C":3}},inplace=True)
    new_data['Embarked'].fillna(new_data['Embarked'].mean(),inplace=True)
    return new_data
dataset.dropna
data=preprocess_data(dataset)
data['Age'] = data['Age'].astype(int)
data['Embarked'] = data['Embarked'].astype(int)
x=data.drop(["Survived", "Name", "Cabin", "Ticket", "Embarked"], axis=1)
y= data["Survived"]

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)

#applying tree algorithm
tree = tree.DecisionTreeClassifier()  
tree.fit(x_train, y_train)   #fitting our model
y_pred=tree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```
