```python
x=df.drop(["target","flower_name"],axis=1)
y=df["target"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2 ,random_state=1)
print(len(x_train),len(x_test))
```

```python
model=KNeighborsClassifier(n_neighbors=11)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred=model.predict(x_test)
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#apriori now
data=pd.read_csv('Market_Basket_Optimisation.csv', sep=',', header=None)
data.head()
```

```python
data.shape
```

```python
#Putting each customer's shopping items in a list

data2=[]
for i in range(0, 7501):
    data2.append([str(data.values[i,j]) for j in range(0, 20)]) 
#Eliminating nan from our list
data3=[]
temp=[]

for i in range(0, 7501):
    for j in range(0, 20):
        if(data2[i][j]!='nan'):
            temp.append(data2[i][j])
    data3.append(temp)
    temp=[]
data3=np.array(data3,dtype=object)
te=TransactionEncoder()
te_ary=te.fit(data3).transform(data3)    #Apply one-hot-encoding on our dataset
data=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
data
```

```python
frequent_itemsets=apriori(data, min_support=0.05, use_colnames=True) #Let’s select itemsets with a minimum of 5% support
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate itemsets with confidence over 20%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.01)
```
